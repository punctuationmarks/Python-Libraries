### Stack plots
(Kinda what it sounds like, stacking plots on top of each other like layers of a cake)

````
# generating data
days = [1, 2, 3, 4, 5, 6, 7]

# in hours
sleeping = np.random.randint(5, 10, size= 7)
working = np.random.randint(5, 13, size= 7)
playing = np.random.randint(1, 3, size= 7)
# hours spent eating/cooking/buying food
eating = np.random.randint(1, 4, size= 7)

# The elusive "lost time" where we thought we recorded everything, 
  # but apparently not
# total hours in the week - all the activities during the week
lt_1 = 24 - (sleeping[0] + working[0] + playing[0] + eating[0])
lt_2 = 24 - (sleeping[1] + working[1] + playing[1] + eating[1])
lt_3 = 24 - (sleeping[2] + working[2] + playing[2] + eating[2])
lt_4 = 24 - (sleeping[3] + working[3] + playing[3] + eating[3])
lt_5 = 24 - (sleeping[4] + working[4] + playing[4] + eating[4])
lt_6 = 24 - (sleeping[5] + working[5] + playing[5] + eating[5])
lt_7 = 24 - (sleeping[6] + working[6] + playing[6] + eating[6])
lost_time = [lt_1, lt_2, lt_3, lt_4, lt_5, lt_6, lt_7]
````

````
# x and then all of the y values "stacked" on top of each other
plt.stackplot(days, sleeping, working, playing, eating, lost_time)

plt.xlabel("Day")
plt.ylabel("Hours")
plt.title('Stack Plots')
plt.show()
````

### Adding colors (Applies to all types of visual displays)
Matplotlib has these color options and more at [Matplotlib Colors](https://matplotlib.org/api/colors_api.html):



1.  RGB or RGBA tuple of float values between 0 and 1:
      *   so if you had the RGB color of (255, 150, 255) which is a pinksh color, 
            divide each of the numbers by 255 (which is the max value of each color channel)
            to get the 0 to 1 value (the percentage value of 100% represented by 1)
      *   so that tuple would be (1, 0.588, 1)


2.   Hex RGB or RGBA as a string
       * So that'd look like '#0F0F0F' or '#0F0F0F0F'


3. A string reprepresentation of a number between 0 and 1 to be JUST a variation of the gray color
      * So '1' would be dark gray, '0' woud be light gray, and '0.5' would be in between with every variations in between that

4. A string version of a X11/CSS4 color name [Wiki on X11 colors](https://en.wikipedia.org/wiki/X11_color_names)

5. You can also just use the shorthand of 'CYMK' ( cyan, magenta, yellow, and key) as 'c', 'y', 'm', 'k'

6. And more if you need to get specific, go here: [Matplotlib Colors](https://matplotlib.org/api/colors_api.html)

### Subplots
In Matplotlib the "figure" is implied, but if you want to modify it you have to reference it


Once you have a subplot, you can plot things on that subplot directly



Good way to organization subplots (you must at least call the figure and axis and modify those before building the main plot): 

1.   Reference Figure
2.   Reference and define your axis (subplots)
3.   Set data and plot data onto corresponding subplots and main graph
     * Define the data by having it organized here (it might be too much to have it imported from online and cleaned here, try to use clean techniques)
4.   Define any modifications to subplots (ax1, ax2, ax3, ect)
     * Altering label rotation, color, size, ect of each individual subplot 
5. Build main plot
     * Here you will define the main xlabel and ylabel, title, legend, and show



#### One Subplot for customization

````

# referencing the figure of the graph itself
fig = plt.figure()

# the first reference to a subplot, 
  # the first tuple is the size of the grid 
    # (meaning the size of the measurement of the subplot graph, 
    # so a (1, 1) visualizes the (x, y) by 1 each 
    # and a (5, 5) visualizes the (x, y) by 5 each)
  # the second tuple is the starting point of the plot
ax1 = plt.subplot2grid((1, 1), (0,0))



# loading some pseudo random data
x_line_a, y_line_a = [1, 2, 3, 4], [10, 15, 9, 25]

# adding a second line 
x_line_b, y_line_b = [1.0, 1.5, 2.0, 4.5], [10, 16, 18, 20]



# since we're plotting these on the subplot, 
# we plot them directly onto the subplt
  # NOTE, it will take up as much space as the other plots, 
  # but it's just seperate as a subplot
ax1.plot(x_line_a, y_line_a, label="Line A")
ax1.plot(x_line_b, y_line_b, label="Line B")


# want to set the ticker labels under your subplot on a tilt?
# you can loop over each of the tickers and adjust them accordingly
for label in ax1.xaxis.get_ticklabels():
  label.set_rotation(45)

# but we still need to make a main plt 
# (which the subplot goes on top of)
plt.xlabel('X Label on Main plt')
plt.ylabel('Y Label on Mian plt')
plt.title('Using a subplot')
plt.legend()
plt.show()
````

#### Multiple Subplots

````

# referencing the figure of the graph itself
fig = plt.figure()

# the first reference to a subplot, 
  # the first tuple is the size of the grid 
    # (meaning the size of the measurement of the subplot graph, 
    # so a (1, 1) visualizes the (x, y) by 1 each 
    # and a (5, 5) visualizes the (x, y) by 5 each)
  # the second tuple is the starting point of the plot
ax1 = plt.subplot2grid((1, 1), (0,0))

ax2 =  plt.subplot2grid((1, 1), (0,0))



# loading some pseudo random data
x_line_a1, y_line_a1 = [1, 2, 3, 4], [10, 15, 9, 25]
x_line_b1, y_line_b1 = [1.0, 1.5, 2.0, 4.5], [10, 16, 18, 20]
x_line_c1, y_line_c1 = [1, 2, 10, 4], [8, 14, 9, 15]
x_line_d1, y_line_d1 = [1.5, 1.5, 2.0, 4.5], [10, 16, 18, 20]

x_line_e1, y_line_e1 = [2, 1.2, 2.2, 4.2], [12, 14, 20, 22]
x_line_f1, y_line_f1 = [1, 1.2, 2, 4], [13, 15, 21, 23]


# lines a1, b1, c1, and d1 are all on ax1
ax1.plot(x_line_a1, y_line_a1, label="Line A1")
ax1.plot(x_line_b1, y_line_b1, label="Line B2")
ax1.plot(x_line_c1, x_line_c1, label="Line C1")
ax1.plot(x_line_d1, y_line_d1, label="Line D1")

# lines e1 and f1 are on ax2

ax2.plot(x_line_e1, y_line_e1, label="Line E1")
ax2.plot(x_line_f1, y_line_f1, label="Line F1")

# gives the backdrop of the subplot a grid, this is customizeable 
ax1.grid(True)


# want to set the ticker labels under your subplot on a tilt?
# you can loop over each of the tickers and adjust them accordingly
for label in ax1.xaxis.get_ticklabels():
  label.set_rotation(45)

  
  
for label in ax2.xaxis.get_ticklabels():
  label.set_rotation(20)
  
  
# but we still need to make a main plt 
# (which the subplot goes on top of)
plt.xlabel('X Label on Main plt')
plt.ylabel('Y Label on Mian plt')
plt.title('Using multiple subplots')
plt.legend()

# adjusting the view of the multiple subplots
plt.subplots_adjust(left=0.09, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.2)
plt.show()
````

````
# having some randomly generated data
uniform_skewed_1 = np.random.rand(100) * 100 - 40
high_outliers_1 = np.random.rand(10) * 50 + 100
low_outliers_1 = np.random.rand(10) * -50 - 100

uniform_skewed_2 = np.random.rand(80) * 100 - 25
high_outliers_2 = np.random.rand(10) * 50 + 100
low_outliers_2 = np.random.rand(10) * -50 - 100

uniform_skewed_3 = np.random.rand(70) * 100 - 40
high_outliers_3 = np.random.rand(10) * 50 + 100
low_outliers_3 = np.random.rand(10) * -50 - 100

uniform_skewed_4 = np.random.rand(75) * 100 - 20
high_outliers_4 = np.random.rand(10) * 50 + 100
low_outliers_4 = np.random.rand(10) * -50 - 100

uniform_skewed_5 = np.random.rand(75) * 100 - 15
high_outliers_5 = np.random.rand(10) * 50 + 15
low_outliers_5 = np.random.rand(10) * -50 - 25

# organizing the data into candle sticks
data_1 = np.concatenate((uniform_skewed_1, high_outliers_1, low_outliers_1))
data_2 = np.concatenate((uniform_skewed_2, high_outliers_2, low_outliers_2))
data_3 = np.concatenate((uniform_skewed_3, high_outliers_3, low_outliers_3))
data_4 = np.concatenate((uniform_skewed_4, high_outliers_4, low_outliers_4))
data_5 = np.concatenate((uniform_skewed_5, high_outliers_5, low_outliers_5))
````

````
fig, ax = plt.subplots()

# plotting each boxplot on seperate areas to ensure they won't overlap
  # Note: you have to declair fig, ax if you want to have multiple boxplots
  # unless your intended goal is to have them overlap
  
# widths is how wide the boxes will be
  
# having a "notch" in the median of the box
ax.boxplot(data_1, 
            positions=[1], 
            notch=True, 
            widths=0.5)

# changing the outpoint symbols (gD <- green diamond)
ax.boxplot(data_2,                     
            positions=[2], 
            sym="gD", 
            widths=0.25)

# changing the whisker lengths
# changing the outpoint symbols (ks <- black square)
ax.boxplot(data_3,                      
            positions=[3], 
            notch=True,
            sym="ks",
            whis = 0.5,
            widths=0.25)


## Adding Colors to boxplots:
  #Two Options:
    # 1. Specify Props Dictionaries inside the ax.boxplot()
    # Lots to unpack here: 
        # boxprops 
            # facecolor the bulk/body/face of the box
            # color is the outline of the box
        # capprops
            # the line that shows the highs and lows
        # whiskerprops
            # the lines that extend out of the box (the whiskers themselves)
        # Medianprops
            # the color of the line at the median, where the notch would be

ax.boxplot(data_4, positions=[4], patch_artist=True, widths=0.80,
            boxprops=dict(facecolor="black", color="red"),
            capprops=dict(color="yellow"),
            whiskerprops=dict(color="green"),
            flierprops=dict(color="brown", markeredgecolor="black"),
            medianprops=dict(color="blue"))


    # 2. Set all of the colors individually outside of the ax.boxplot()
       # This might look a little better, 
       # but you'd have to remember which ['item']
       # took just color and which took color + facecolor or whatever
       # would be better if you wanted more of a uniform look instead of super 
       # custom like ax.boxplot(data_4)
boxplot_5 = ax.boxplot(data_5,                       
                      positions=[5], 
                      notch=True, 
                      patch_artist=True,
                      widths=0.80)

boxplot_5_color = "#ffcc66"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_5[item], color=boxplot_5_color)
plt.setp(boxplot_5["boxes"], facecolor=boxplot_5_color)
plt.setp(boxplot_5["fliers"], markeredgecolor=boxplot_5_color)


# one thing super nice about declairing each boxplot individually as a variable
# is that you can use the color in the legend
ax.legend(['1', '2', '3', '4', '5'], loc='upper right')



# setting the x limit 
ax.set_xlim(0, 6)

ax.set_title('Overly Custom Candle Stick Chart (aka boxplot())')

plt.show()
````

````
# Want to have your legend reflect the color codes of each indiviudal box?
# Make each of them variables, then in the legend grab the ['boxes'][0] color
# from each of them. 

fig, ax = plt.subplots()


boxplot_1 = ax.boxplot(data_1, 
            positions=[1], 
            notch=True, 
            widths=0.5)

boxplot_1_color = "#993333"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_1[item], color=boxplot_1_color)




boxplot_2 = ax.boxplot(data_2,                     
            positions=[2],
            widths=0.25)

boxplot_2_color = "#336600"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_2[item], color=boxplot_2_color)




boxplot_3 = ax.boxplot(data_3,                      
            positions=[3], 
            notch=True,
            whis = 0.5,
            widths=0.25)

boxplot_3_color = "#cc66ff"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_5[item], color=boxplot_3_color)





boxplot_4 = ax.boxplot(data_4, 
                       positions=[4], 
                       patch_artist=True, 
                       widths=0.80)

boxplot_4_color = "#6699ff"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_4[item], color=boxplot_4_color)





boxplot_5 = ax.boxplot(data_5,                       
                      positions=[5], 
                      notch=True, 
                      patch_artist=True,
                      widths=0.80)


boxplot_5_color = "#ffcc66"
for item in ['boxes','whiskers', 'fliers', 'medians', 'caps']:
        plt.setp(boxplot_5[item], color=boxplot_5_color)



ax.legend([boxplot_1["boxes"][0], 
           boxplot_2["boxes"][0],
           boxplot_3["boxes"][0], 
           boxplot_4["boxes"][0], 
           boxplot_5["boxes"][0]], 
          ['1', '2', '3', '4', '5'], 
          loc='upper right')



# setting the x limit 
ax.set_xlim(0, 6)
ax.set_title('Legend matching the Boxes')
plt.show()
````

### Customizing Date Plots

````

# We'll be grabbing data for Stock Market Confidence Indices from Yale
SMCI_Yale =  quandl.get("YALE/US_CONF_INDEX_VAL_INDIV", authtoken=api_key)

````

````
# Let's see what we're working with
SMCI_Yale.head()

````

````
# grabbinng the dates
SMCI_Yale_dates = SMCI_Yale.index

# grabbing the closing price
SMCI_Yale_index = SMCI_Yale['Index Value'] 

SMCI_Yale_standard_error = SMCI_Yale['Standard Error'] 
````

````
fig, ax = plt.subplots()





# this fills everything below the line
    # synatx is 
      # fill_between(x_value, y_value, 
        #  fills down to or up to y==n depending on if xy_line > or < n respectively, 
        #  alpha= zero to 1 transparency)
        
        
# for this example, we're filling in whatever is greater than
# the starting index

ax.fill_between(SMCI_Yale_dates, 
                SMCI_Yale_index, 
                SMCI_Yale_index[0],where=(SMCI_Yale_index >= SMCI_Yale_index[0]),
                facecolor="black",
                alpha=0.5)

# filling in whatever is less than
# the starting index
ax.fill_between(SMCI_Yale_dates, 
                SMCI_Yale_index, 
                SMCI_Yale_index[0],where=(SMCI_Yale_index <= SMCI_Yale_index[0]),
                facecolor="red",
                alpha=0.5)

# is the fillbetween not dynamic looking enough?
# you can add an axis horizontal line 
ax.axhline(SMCI_Yale_index[0], color="black", linewidth=2)




# plotting Index
ax.plot_date(SMCI_Yale_dates, SMCI_Yale_index, fmt='-', label="Index")


# adding the standard error as a gray line
ax.plot_date(SMCI_Yale_dates,SMCI_Yale_standard_error, 
             fmt='-', 
             color="gray",
             label="Standard Error")






ax.legend()

plt.xlabel('Date')
plt.ylabel("$")
plt.title('Stock Market Confidence Indices from Yale')


plt.show()
````

````
# Random data
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]


# nrows is how many rows of graphs you will have
# ncols is how many columns you'll have
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))


data1 = axes[0].plot(all_data)

axes[0].set_title('Data1')

# notch shape box plot
data2 = axes[1].plot(all_data * 2)

axes[1].set_title('Data2')

# adding horizontal grid lines, and x and y labels 
    # (you could also do if you want each to have custom labels, 
    # just like the titles)
      # axes[0].set_xlabel('X Label 1')
      # axes[1].set_xlabel('X Label 2')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xlabel('X Label for both of the graphs')
    ax.set_ylabel('Y Labels for both graphs')

plt.show()
````
